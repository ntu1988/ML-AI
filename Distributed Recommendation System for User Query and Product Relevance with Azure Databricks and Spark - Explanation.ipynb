##### 準備training data and test data

```python
from pyspark.sql.functions import when

# Map string labels to numeric values
train_df = train_df.withColumn(
    "label",
    when(train_df.label == "Exact", 3.0)
    .when(train_df.label == "Partial", 2.0)
    .when(train_df.label == "Irrelevant", 0.0)
    .cast("float")  # Ensure the label column is numeric
)

from pyspark.sql.functions import when

# Map string labels to numeric values
test_df = test_df.withColumn(
    "label",
    when(test_df.label == "Exact", 3.0)
    .when(test_df.label == "Partial", 2.0)
    .when(test_df.label == "Irrelevant", 0.0)
    .cast("float")  # Ensure the label column is numeric
)
```

```python
train_df = train_df.repartition(4)
test_df = test_df.repartition(4)
```

#### Use SparkXGBRanker作為model
Training Process: 一個query會對應到多個product，然後label會告訴model，相關性高還是低，也就是說這個product's product_query_rating_feature相關性是exact, partial, irrelevant，以此來訓練model什麼樣的query應該對應到什麼樣的features從而找出最適合推薦的product.

ranker的設定：1. group讓ranker知道有多少個候選產品在相同的query. 2. query_id是讓我們把對應的productA, ProductB, ProductC的產品拉出來訓練model

pairwise: 在一個分組(query_id)內，生成所有的排序對。

樹的分裂：根據什麼來分裂，來區隔不同label的product

```python
# Train the ranking model using Spark XGBoost
from xgboost.spark import SparkXGBRanker

xgb_ranker = SparkXGBRanker(
    features_col="features",
    label_col="label",
    group_col="group",  # Specify the group column for ranking
    qid_col="query_id",  # Specify the query ID column for ranking
    objective="rank:pairwise",  # Objective function for ranking
    numRound=10,
    maxDepth=5,
    eta=0.1,
    gamma=0.1,
    num_workers=4,
)


ranker_model = xgb_ranker.fit(train_df)
```

```python
# Evaluate the model with MAP
predictions = ranker_model.transform(test_df)
```

##### 預測的值：The ranking scores output by XGBoostRanker are relative values.
Negative scores don't carry any specific meaning by themselves—they're just numerical outputs that can be used to determine the relative order of items.

```python
predictions_pandas = predictions.limit(50).withColumn('product_description_tokens', predictions['product_description_tokens'].cast('string')).withColumn('query_tokens', predictions['query_tokens'].cast('string')).toPandas()
display(predictions_pandas.head(50))
```

##### Evaluate model透過比較y_score給的相對相關性和y_true的相對相關性，來計算average precision

AP 是一個排序評估指標：它衡量模型是否能將相關項目排在不相關項目之前。分數越接近 1.0，模型排序的效果越好。

```python
from sklearn.metrics import average_precision_score
import numpy as np
# Extract labels and predictions
y_true = predictions.select("label").rdd.flatMap(lambda x: x).collect()
y_score = predictions.select("prediction").rdd.flatMap(lambda x: x).collect()
y_score = np.array(y_score).reshape(-1, 1)
# Calculate Average Precision Score
ap_score = average_precision_score(y_true, y_score)
print(f"Average Precision Score: {ap_score}")
```
